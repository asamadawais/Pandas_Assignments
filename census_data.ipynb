```python
import numpy as np
import pandas as pd
from os import listdir
import matplotlib.pyplot as plt
filepaths = [f for f in listdir("./") if f.endswith('.csv')]
usa_census = pd.concat(map(pd.read_csv, filepaths), ignore_index = True)
```

```python
usa_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
usa_census[['Male','Female']] = usa_census.GenderPop.str.split("_",expand=True) 
```

```python
usa_census.drop(columns = ['GenderPop'], inplace=True)
```

```python
usa_census['Income'] = usa_census.Income.str.strip('$')
```

```python
usa_census = usa_census.replace('%','', regex=True)
```

```python
usa_census = usa_census.replace('F','', regex=True)
```

```python
usa_census = usa_census.replace('M','', regex=True)
```

```python
print(usa_census)
```

```python
usa_census.loc[:,'Hispanic':'Income'] = round(usa_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
usa_census['Male'] = usa_census['Male'].astype(int)
```

```python
usa_census.drop(columns = ['Female'], inplace=True)
```

```python
usa_census['Female'] = usa_census['TotalPop'] - usa_census['Male']
```

```python
usa_census.dtypes
```

```python
usa_census.duplicated()
```

```python
usa_census.drop_duplicates(inplace= True)
```

```python
plt.scatter(usa_census['Female'],usa_census['Income'])
plt.xlabel('Salary in US Dollars', fontsize=12)
plt.ylabel('Female Population', fontsize=12)
plt.show()
```
